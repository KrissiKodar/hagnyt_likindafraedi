## L14.1 The Ito Integral
​
Recall that the Riemann Integral is defined as the limit of a sum (https://en.wikipedia.org/wiki/Riemann_integral):

$$\int_a^b{f(x)dx} = \lim_{n \rightarrow \infty}\sum_{i=0}^{n-1}{f(x_i)(x_{i+1}-x_i)}$$

The interpretation of the integral is the area under the curve of the function $f(x)$. The Riemann integral has a generalization called the Riemann-Stieltjes integral. If $G$ is a function of bounded variation (e.g., continuously differentiable - more details [here])(https://en.wikipedia.org/wiki/Bounded_variation), then we can define the Riemann-Stieltjes integral of g with respect to G as

$$\int_a^b{g(x)dG} = \lim_{n \rightarrow \infty}\sum_{i=0}^{n-1}{g(x_i)(G(x_{i+1})-G(x_i))}$$

If $G$ is differentiable, then the Riemann-Stieltjes integral reduces to

$$\int_a^b{g(x)dG} = \int_a^b{f(x)G'(x)dx}$$

We will define the Ito integral in an analogous way as the limit

$$\int_0^t{X(s,\omega)dW(s,\omega)} = \lim_{n \rightarrow \infty}\sum_{i=0}^{n-1}{X(s_i,\omega)(W(x_{i+1},\omega)-W(x_i,\omega))}$$

where $X(s)$ is assumed to be $F_t$ adapted, where $F_t$ is the filtration generated by the Brownian motion. This means that if we know the full path of the Brownian motion up to time $s$, then we know what $X(s)$ is. We also require that

$$\int_0^t{E[X^2(s)]ds} < \infty$$

To see why we need this criterion, we recall that

$$E[(W(s_{i+1}) - W(s_i))^2] = s_{i+1} - s_i$$

which comes directly from the properties of the Brownian motion. Since $X(s)$ is adapted to $F_s$, we know that  $X(s_i)$ is independent of $W(s_{i+1}) - W(s_i)$ (as it is in the future of $X(s_i)$), so

$$E[\left(X(s_i)(W(s_{i+1}) - W(s_i))\right)^2] =  E[X(s_i)^2](s_{i+1} - s_i)$$

and so

$$E\left[\left(\sum_{i=0}^{n-1}{X(s_i)(W(s_{i+1}) - W(s_i))}\right)^2\right] =  \sum_{i=0}^{n-1}{E[X(s_i)^2](s_{i+1} - s_i)}$$

because all the variables under the summation sign are independent. Now, if we take the limits as n goes to infinity, we see that the right hand side is just

$$\int_0^t{E[X^2(s)]ds}$$

while the left hand side contains the definition of the stochastic integral. If we require the limit of the right hand side to be finite, then, we can define the left hand side, the Ito integral. This is in fact a theorem called the Ito isometry:

$$E\left[\left(\int_0^t{X(s)dW}\right)^2\right] =  \int_0^t{E[X^2(s)]ds}$$

Here are some properties of the Ito integral:

$$\int_0^t{X(s)dW} = \int_0^s{X(s)dW} + \int_s^t{X(s)dW}$$

$$\int_0^t{aX(s) + bY(s)dW} = a\int_0^t{X(s)dW} + b\int_0^t{Y(s)dW}$$

$$E\left[\int_0^t{X(s)dW}\right] = 0$$

and $\int_0^t{X(s)dW}$ is a stochastic process itself that is $F_t$ adapted (i.e., if we know the story of $W(t)$ up to time $t$, then we know the value of the Ito integral). Furthermore, the Ito integral is a martingale with respect to $F_t$.

All of these are clearly true for the finite sums and extend to the integral by taking the limit.

Let's do a simple example. What is $\int_0^t{adW}$ where $a$ is a constant? 

Let's look at the finite sums where the interval from 0 to $t$ is split into $n$ pieces:

$$\sum_{i=0}^{n-1}{a(W(s_{i+1}) - W(s_i))} = a(W(t) - W(0)) = aW(t)$$

since the sums telescope. Taking the limit reveals the same answer. Note that this integral is indeed a $F_t$ adapted process.

Definition is not very good to perform integration, but we have a tool called the Ito formula that helps us finding closed formulas for some Ito integrals.

## L14.2 The Ito Formula or Ito Lemma

The Ito formula (sometimes called the Ito lemma) is useful for actually calculating integrals. We start with the definition of an Ito process, or a stochastic integral:

$$X_t = X_0 + \int_0^t{\mu_sds} + \int_0^t{\sigma_sdW_s}$$

where $\mu_s$ and $\sigma_s$ are stochastic process. $\sigma_s$ must be squarely integrable (i.e., the integral of $E[\sigma_s^2]$ over the fixed time interval must be finite), and $\mu_s$ should also be integrable. This process is then written in a shorter and more handy way in the following differential form:

$$dX_t = \mu_tdt+\sigma_tdW_t$$

We will be using the differential form of an Ito process quite a bit going forward.

A simple example when 

$$X_t = \mu t + \sigma W_t$$

where $\mu$ and $\sigma$ are constants, then we have

$$dX_t = \mu dt+\sigma dW_t.$$

What if we have this stochastic differential equation?

$$dX_t = X_t(\mu dt+\sigma dW_t).$$

Things start to get more difficult, so we need a tool...

We now move to the important Ito formula. If $X_t$ is an Ito process with

$$dX_t = \mu_tdt+\sigma_tdW_t$$

and $g(t,x)$ is a twice continuously differentiable function, then

$$Y_t = g(t,X_t)$$

is also an Ito process and

$$dY_t = \frac{\partial g}{\partial t}(t,X_t) dt + \frac{\partial g}{\partial x}(t,X_t)dX_t + \frac{1}{2}\frac{\partial^2 g}{\partial x^2}(t,X_t)(dX_t)^2$$

where we use the following rules:

* $dtdt= 0$
* $dtdW_t= 0$ 
* $dW_tdW_t=dt$

Let's see a quick example before we try to understand why the formula is true. 

If $Y_t = W_t^2/2$ then we take $g(t,X_t) = X_t^2/2$. In this example, $X_t = W_t$ since $Y_t = g(t,W_t)$, but in general, $X_t$ can be any Ito process. The partial derivatives of $g$ are:

$$\frac{\partial g}{\partial t}(t,X_t) = 0$$

$$\frac{\partial g}{\partial x}(t,X_t) = X_t$$

$$\frac{\partial^2l g}{\partial x^2}(t,X_t) = 1$$

Inserting into the Ito formula gives:

$$dY_t = 0dt + W_tdW_t +  \frac{1}{2}1(dW_t)^2 = W_tdW_t + \frac{1}{2}dt$$

What does this mean in terms of the Ito integral?

$$Y_t = W_t^2/2 = \int_0^t{\frac{1}{2}dt} + \int_0^t{W_tdW_t}$$

Re-arranging, gives is

$$\int_0^t{W_tdW_t} = \frac{W_t^2}{2} -\frac{t}{2}$$ 

**Exercise**: use Ito's formula to show that $d(tW_t) = W_tdt+tdW_t$

Why is the Ito formula true? Let's use the Taylors theorem to see what is going on.

Let's consider a function of the Brownian motion only to keep it simple:

$$f(W_{s_{i+1}}) - f(W_{s_i}) = f'(W_{s_i})(W_{s_{i+1}} - W_{s_i}) + \frac{1}{2}f''(W_{s_i})(W_{s_{i+1}} - W_{s_i})^2 + \text{hot}$$

The term on the left is just a representation of $df$, the first term on the right is a representation of $f'(W_s)dW$, and the last term is a representation of $\frac{1}{2}f''(W_s)dW_t^2$. We know that $E[W_t^2] = t$, and that leads to $dW_t^2 = dt$. Of course, all of this can be proved formally, but it is a bit elabotate. Higher order terms (hot) lead to $dt^{3/2}$ plus higher orders that are negligable compared to $dt$. 

If we write out what $dX_t$ is in the Ito formula, we geta bit more messier formula:

$$dY_t = \left[\frac{\partial g}{\partial t}(t,X_t) +\mu \frac{\partial g}{\partial x}(t,X_t) + \frac{1}{2}\sigma \frac{\partial^2 g}{\partial x^2} \right]dt + \sigma \frac{\partial g}{\partial x}(t,X_t)dW_t$$

One of the workhourses that we will be using is the Geometric Brownian Motion, defined by the stchastic differential equation:

$$\frac{dX_t}{X_t} = \mu dt + \sigma dW_t$$

where $\mu$ and $\sigma$ are constants. We use this process to describe, e.g., a stock price.

In order to solve this equation, let's consider $g(x) = \ln(x)$ and apply the Ito formula:

$$d(ln(X_t)) = 0dt + \frac{dX_t}{X_t} - \frac{1}{2}\frac{(dX_t)^2}{X_t^2} = \frac{dX_t}{X_t} - \frac{1}{2}\sigma^2 dt$$

or

$$\frac{dX_t}{X_t} = d(ln(X_t)) + \frac{1}{2}\sigma dt$$

We also have that

$$\frac{dX_t}{X_t} = \mu dt + \sigma dW_t$$

and combining these things together, we get

$$ \mu dt + \sigma dW_t= d(ln(X_t)) + \frac{1}{2}\sigma^2 dt$$

or

$$d(ln(X_t)) = (\mu - \frac{1}{2}\sigma^2) dt + \sigma dW_t.$$

Finally,

$$X_t = X_0 \exp\left( (\mu - \frac{1}{2}\sigma^2)t+ \sigma W_t\right)$$


Let's take a look at one sample path of the geometric brownian motion. If $t$ is in years, we could e.g., assume that $\mu$ = 10% and $\sigma=20%$.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

dt = 1 / 252.0  # Business days in a year
mu = 0.1
sigma = 0.2

initial_price = 100
# Vectorised implementation of asset path generation
# including four prices per day, used to create OHLC
asset_path = np.exp((mu - sigma**2 / 2) * dt +
            sigma * np.random.normal(0, np.sqrt(dt), size=(252)))

path = initial_price*asset_path.cumprod()
plt.plot(range(252),path)
```

We can also re-arrange the stochastic differential equation of a geometric differential equation. If we consider

$$\frac{dX_t}{X_t} = (\mu - \frac{1}{2}\sigma^2) dt + \sigma dW_t$$

then the solution to that equation is

$$X_t = X_0 \exp\left( \mu t+ \sigma W_t\right),$$

a form that is sometimes useful.

Lastly, to close out this discussion on stochastic differential equations, let me mention that under some restrictions on the $\mu$ and $\sigma$ processes, there exists a unique continuous solution to the stochastic differential equation

$$dX_t = \mu(t,X_t) dt+\sigma(t,X_t) dW_t.$$


## L14.3 Martingale representation theorem

We saw before that an Ito Integral is a martingale, but the reverse is also true. Any martingale with respect to the filtration $F_t$, generated by $W_t$ can be expressed as an Ito integral. This is a very important theorem when it comes to pricing of derivatives. The story is, essentially, that we will change our underlying probability measure and convert the value of a financial derivative into a martingale. Then, using the martingale representation theorem, we can derive a stochastic differential equation for the price, which we can then solve and get a formula for the derivative price.

The theorem states that: Suppose that $M_t$ is a square integrable martingale with respect to $F_t$, then there exists a unique stochastic process $g(s,\omega)$ such that

$$M_t(\omega) = E[M_0] + \int_0^t{g(s,\omega)dW_s}$$

A more relevant representation theorem is the following:

If $M_t$ is a martingale with a non-zero volatility (i.e., it is not a deterministic process), then if $N_t$ is any other martingale, then there exists an $F_t$ adapted process $\phi_t$ such that

$$N_t = N_0 + \int_0^t{\phi_sdM_s}$$

and by $\int_0^t{\phi_sdM_s}$ we mean that we expand the differential and carry out two integrals, one deterministic and one stochastic.
