## L13.1 Filtrations

The concept of filtration is a bit abstract, but the most important point of it is that it contains all information up to a given point in time (we know what has happened).

We start with a sigma algebra, which defines all allowable events:

A $\sigma$ algebra $F$ on $\Omega$ is a family of subsets on $\Omega$ such that
* the empty set $\emptyset$ belongs to $F$
* if A belongs to $F$, then $A^c$ belongs to $F$
* if $A_1, A_2$, ..., is a sequence of sets in $F$, then their union $A_1 \cup A_2 \cup$ ... also belongs to $F$.

These are all allowable events. So far we have assumed that all possible subsets belong $F$, which is the case when we have finite $\Omega$, and we can still assume, for practical purposes, that all subsets belong to $F$.

A *filtration* is a sequence of $\sigma$ fields such that

$$F_1 \subset F_2 \subset ... \subset F_n \subset ...$$

The important concept in all of this abstraction is that $F_n$ represents the knowledge at time $n$. It contains all events $A$ such that at time $n$ it is possible to decide whether $A$ has occured or not. As $n$ increases there will be more and more such events so that our knowledge becomes larger and larger. The longer you live the wiser you become (at least for some people).

Let's look at some examples:

A = {the first 5 tosses produce at least 2 heads}

At time $n=5$, i.e., once the coin has been tossed 5 times, it will be possible to decide whether A has occured or not, so that $A \in F_5$. It is not *always* possible to decide whether A has occured or not when we have tossed the coin 4 times, so $A \notin F_4$. It is sometimes possible, but the requirement is that must be possible to determine whether A has occured or not in *all* situations.

B = {the first 100 tosses produce the same outcome}

C = {there is at least one head in the sequence $X_1$, $X_2$, ...}

A sequence of random variables $X_1, X_2, ...$  is *adapted* to the filteration $F_1, F_2, ...$ if $X_n$ is measurable with respect to $F_n$

Measurable here can be taken to mean that once we reach the time $n$, we know the value of $X_n$ if the random variables are adapted to the filteration.

## L13.2 Martingales

Martingales is a stochastic process $X_1, X_2, ...$ adapted to a filtration so that

$$E[X_{n+1}|F_n] = X_n$$

In other words, martingales are stochastic processes where the expected value of the next step, given the information up to the current time, is the current value. Martingale thus represents a fair game.

Example: random walk. Flip a coint and get 1 and -1 with probabilities 0.5, and define $S_n = \sum_{k=1}^n{X_k}$

Let's show that $Y_n = S_n^2 - n$ is a martingale. This is closely related to Brownian Motion. The key is to transform $S_{n+1}^2 = (S_n+X_{n+1})^2 = S_n^2 + 2S_nX_{n+1} + X_{n+1}^2$

Another example is $X_n = E[X|F_n]$, which is also a martingale - current expected value of a random variable.

It does not really matter how far into the future you go. If $X_n$ is a martingale, then

$$E[X_{n+i}|F_n] = X_n$$

for all i>0. 

Martingales relate to fair games. Suppose $X_1, X_2, ..., X_n$ represent your winnings or losses per unit stake in $n$ games, and define

$$S_n = X_1 + ... + X_n$$.

Then, the game is fair if

$$E[S_{n+1}|F_n] = S_n$$

i.e., if you have played $n$ games, then your expected earnings after playing the next game is the same as you currently have. I.e., a fair game is a martingale.

A gambling strategy, $\alpha_1, \alpha_2, ...$ is the set of stakes gambled in each game (important that they are adapted to the filtration $F$, i.e., you don't have any informational advantage or in other words no cheating). The total winnings is then

$$U_n = \alpha_1X_1 + ... + \alpha_nX_n = \alpha_1(S_1-S_0) + ... + \alpha_n(S_n-S_{n-1}).$$

There is an interesting result that says that if $S_n$ is a martingale (i.e., the game is fair), then $U_n$ is also a martingale. I.e., you can't beat a fair game using a gambling strategy. This can also be extended to unfair games and favorable games. Unfair games always remain unfair and favorable games will remain favorable.

Martingales are also common in option pricing, and is it possible to show that $h(x,n) = E[(x-k)^+|F_n]$ is a martingale, but we will spend more time on that later on.

## L13.3 Stopping times

One of the nice applications of martingales relate to the concept of stopping times. 

Consider playing a game and assume we have decided to play the game for $\tau = 10$ rounds, no matter what. Then our *stopping time* is 10. In general though, we would like to stop when some condition is satisfied, e.g., if we are broke, or if we have earned some prescribed amount. In this case, the stopping time $\tau$ is a random variable, and it might be interesting to understand more about it - e.g., its expectation or even its distribution. We can use simulations to estimate stopping times under very general conditions, but there are also interesting theoretical results for martingales. 

First, we define $\tau$ to be a stopping time with respect to a filtration if ${\tau = n} \in F_n$, i.e., if at time $n$, we have enough information to know whether or not we have stopped or not.

An example of a stopping time is:

$$\tau = min\{n: S_n = 0 \text{ or } S_n = 10\}$$ 

The main result is called the **optional stopping theorem**, which says that, under rather general conditions (e.g., if the martingale is bounded):

$$E[X_{\tau}] = E[X_1]$$

Note that if $X_i$ is a martingale, then $E[X_1] = E[X_n]$.

Now, here are some interesting applications:

Let's consider the simple random walk again with $X_i = 1$ with probability 1/2 and $X_i=-1$ with probability 1/2, and the process we are interested in is $S_n = X_1 + ... + X_n$. Now, let's consider the first time when we leave a certain interval (a,b), i.e.,

$$\tau = min\{n: S_n = a \text{ or } S_n = b\}$$ 

If we start at some position s, i.e., $S_0=s$, then the optional stopping theorem says that

$$E[S_{\tau}] = E[S_1] = s$$

Now we use out knowledge of $S_{\tau}$. It must be either $a$ or $b$ at $\tau$, so

$$s = E[S_{\tau}] = ap_a + bp_b$$

The sum of the probabilities must be 1, so

$$p_b = \frac{s-a}{b-a}$$

Let's simplify this a little bit, to save space on the algebra, and assume that $a = 0$ and $b=2s$, i.e., stop if we are ruined or if we have doubled out initial value.

Then, $p_{2s} = 1/2$ and $p_0 = 1/2$, as expected. How long can we expect this to take?

We also know that $S_n^2-n$ is a martingale, so we can apply the optional stopping theorem to get

$$E[S_{\tau}^2-\tau] = E[S_1^2-1]$$

$$E[S_1^2] = 1/2*((s+1)^2 + (s-1)^2) = s^2 + 1$$

$$E[S_{\tau}^2] = a^2p_a + b^2p_b = 2s^2$$

so,

$$E[\tau] = s^2$$

Let's verify these probabilities and expected values using Monte Carlo. Start with a wealth of 5 and count the number of times we go broke or reach 10, and the length of time until we get there.

```python
import numpy as np
import random 

N = 10000
s0 = 5 #Of each component

ruin = 0
doubling = 0
times = []
for i in range(N):
    s = s0
    t = 0
    while 0 < s < 10:
        x = 1 if random.random() < 0.5 else -1
        s += x
        t += 1
        if s == 0:
            ruin += 1
        elif s == 10:
            doubling += 1
    times.append(t)

print("Average stopping time: ", sum(times)/N)
print("Probability of ruin: ", ruin/N)
print("Probability of doubling: ", doubling/N)
```


## L13.4 Brownian Motion

We now move to Brownian motion, which will be key concept for the rest of this course.

A Brownian Motion, or a Wiener Process, $W(t)$ for $t$ in $[0,\infty)$ is a stochastic process that satisfies the following conditions:

1) $W(0) =0$

2) The sample paths, $t \rightarrow W(t)$ are continuous 

3) **Independent increments** Whenever $0 = t_0 < t_1 < ... < t_n$

$W(t_1) - W(t_0), W(t_2) - W(t_1), ..., W(t_n) - W(t_{n-1})$ are independent.

4) **Stationary increments** The distribution of $W(t) - W(s)$ is $N(0,\sigma^2(t-s))$ for $s<t$


Brownian motion is called standard Brownian motion if $\sigma=1$. We will mostly work with standard Brownian motion.

Brownian motion originates with Robert Brown in the 1800s who was studying pollen grains in water under a microscope, and he observed that the grains performed a continuous swarming motion. The grains were colliding with water molecules in a random fashion, basically a sum of many independent random effects. The normal distribution thus comes to play due to the central limit theorem. 

Note that from (4)  $W(t)$ is $N(0,\sigma^2t)$ since $W(0) = 0$.

This means that

$$E[W_t^2] = t$$

for a standard Brownian motion.

Let's take a look a path of a standard Brownian motion. Here we simply use that $W(t) = W(t) - W(t-\delta) + ( W(t-\delta) - W(t-2\delta)) + ... + (W(\delta) - W(0))$, with $\delta=1$, is a sum of standard normal variables by property 4.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
path_size = 100


W = 0
path = [0.0]
for i in range(path_size):
    incr = np.random.randn()
    W += incr
    path.append(W)
    
plt.plot(path)
```

Proving that there is such a process, i.e., that the formalities exist, the sigma algebras for the uncountibly many random variables, etc., is quite technical and we will not worry about that in this course.

One can derive several interesting and useful properties from the definition. 

One useful property is that

$$ E[W_tW_s] = \min(t,s)$$

which can be derived by looking at $E[(W_t - W_s + W_s)W_s]$ assuming $s \le t$.

Another is the fact that $W_t$ is a martingale. To see that, consider

$$E[W_t|W_s] = E[W_t - W_s + W_s|W_s] = E[W_t - W_s|W_s] + E[W_s|W_s] = 0 + W_s = W_s$$

Another commonly occuring expression is also a martingale, namely $W_t^2 - t$:

$$E[W_t^2 - t|W_s] = E[(W_t-W_s+W_s)^2 - t|W_s] = W_s^2 + t - s - t = W_s^2 - s$$

Now, we can apply the optional stopping theorem to find some interesting stopping times.

When does the Brownian motion exit an interval (a,b) where a < 0 and b > 0?

$$E[W_{\tau}] = E[W_1] = 0$$

$$E[W_{\tau}] = ap_a + bp_b$$

and the probabilities sum to 1, so

$$p_a = \frac{b}{b-a}$$

$$p_b = \frac{-a}{b-a}$$

How long can we expect this to take?

Use that $W_t^2 - t$ is a martingale, so

$$0 = E[W_1^2 - 1] = E[W_{\tau}^2 - \tau]$$

$$E[W_{\tau}^2] = p_aa^2 + p_bb^2$$

Simplifying, yields

$$E[\tau] = -ab$$


## L13.5 The Brownian bridge
The Brownian bridge is sometimes useful in practical situations, e.g., when one wants to fill in missing data with something that might be plausible. For example, one might have a gap in price series of a share, or the thermometer broke over a period of time, but you need to have some values in the series to perform the analysis you are working on. The Brownian bridge is a process that ranges from $t=0$ to $t=1$ (but can be scaled to any size), and it is defined as

$$W^0_t = W_t - tW_1$$

We need to know what $W_1$ is, i.e., the future is known. The process starts at 0, $W^0_0 = 0$, and ends at $W^0_1 = W_1 - 1*W_1 = 0$, and is random in between. The Brownian bridge, $W^0_t$ is a Gaussian process with mean 0 and the covariance is $E[W^0_sW^0_t] = s(1-t)$, but the latter can be found by expanding using the definition of the Brownian bridge. The variance is then $t(1-t)$.

We can also scale the Brownian bridge to any length of time, $T$, and the it becomes

$$W^0_t = W_t - \frac{t}{T}W_T$$

Now suppose that observations between 100 and 120 went missing - the data collection system went down and no data was collected during that period. We know that the USD was 129.37 in the 100th observation (index=99 in python), and that the 121st observation was 130.23. Assume that our system needs to have some observations in between (maybe because there are other observations and the USD is not the most critical one, but needs to be present). Then, one way is to simulate a Brownian bridge to close the gap.

In this case, we need a linear bridge between the two observations, and then on top of that, we add a brownian bridge with the same standard deviation, $\sigma$, as what we have observed before the missing data:

$$Y_t = \frac{130.23 - 129.37}{(121-100)}*(t-100) + 129.37 + \sigma(W_{t-100}-\frac{t-100}{21}W_{21})$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

sigma = np.std(USD[80:100])

print("Standard deviation if USD using first 100 observations: ",sigma)

obs_100 = USD[99]
obs_121 = USD[120]

def proxy():
    zero_to_twentyone = np.array([range(0,22)])[0]
    linear_bridge = (obs_121 - obs_100)/21*zero_to_twentyone + obs_100
    
    wiener = np.concatenate((np.array([0]),np.random.normal(0, 1, 21)))
    brownian_bridge = wiener - zero_to_twentyone/21*wiener[-1]
    
    res = linear_bridge + sigma*brownian_bridge
    return res[1:-1] #Only return the points in between and skip the ends (both are zero)

USD_proxy = USD.copy()
p = proxy()
for ind in range(100,120):
    USD_proxy[ind] = p[ind-100]
    
plt.plot(x[90:130],USD[90:130],x[90:130],USD_proxy[90:130],'--')
```

Finally, let me just briefly introduce multidimensional Brownian motion. In this case, we still have independent increments, etc. but at any given point in time, $W_i(t)$ are jointly normally distributed with a covariance matrix $C$. It is standard if all the variances are 1, in which case the covariance matrix is just a correlation matrix. We will look closer into that when we have to.

Here are some interesting facts about Brownian Motion:

* $W$ is continuous everywhere but differentiable nowhere
* $W$ will eventually hit every positive or negative number
* Once $W$ hits a value, it immediately hits that same value infinitely many times
* $W$ is a fractal - if you magnify a graph of a Brownian montion, it will always look the same no matter how much you magnify it
